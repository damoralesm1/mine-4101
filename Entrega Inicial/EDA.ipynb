```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Analisis Exploratorio de Datos (EDA)

```python
# Cargar el archivo CSV
file_path = './data_final.csv'
df = pd.read_csv(file_path)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Convertir la columna "Date" a datetime si existe
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Showing dataframe dimensions
df.shape
```

```python
# Showing column types
df.dtypes
```

```python
# Mostrar la descripci√≥n general de los datos
print("Descripci√≥n general de los datos:")
print(df.describe(include='all'))
```

```python
# Seleccionar solo las columnas con variables num√©ricas
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_columns].head()
```

```python
# Reemplazar NaN por 0 en todas las columnas num√©ricas
df[numerical_columns] = df[numerical_columns].fillna(0)
df[numerical_columns].head()
```

### 2. Analisis Univariado

#### Distribuci√≥n de Costos

```python
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
sns.histplot(df['Cost'], kde=True)
plt.title('Distribuci√≥n de Costos')
```

#### Scatterplot entre Costos y Clicks (all)

```python
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 4)
sns.scatterplot(x='Cost', y='Clicks (all)', data=df)
plt.title('Relaci√≥n entre Costos y Clicks (all)')
```

```python
# Graficar Unique Leads por Campa√±a a lo largo del tiempo
plt.figure(figsize=(12, 8))
sns.lineplot(x='Date', y='Unique leads', hue='Campaign name', data=df)
plt.title('Unique Leads por Campa√±a a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Unique Leads')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=1)
plt.tight_layout()
plt.show()
```

```python
# Filtrar solo las filas donde 'Unique leads' es mayor a 0
df_non_zero_leads = df[df['Unique leads'] > 0]

# Obtener las 5 campa√±as con m√°s registros donde 'Unique leads' es mayor a 0
top_campaigns_non_zero = df_non_zero_leads['Campaign name'].value_counts().head(5).index

# Filtrar el DataFrame para incluir solo las 5 campa√±as con m√°s registros con Unique leads > 0
df_filtered_non_zero = df_non_zero_leads[df_non_zero_leads['Campaign name'].isin(top_campaigns_non_zero)]

# Graficar Unique Leads para las 5 campa√±as con m√°s registros (donde Unique leads > 0) a lo largo del tiempo
plt.figure(figsize=(12, 8))
sns.lineplot(x='Date', y='Unique leads', hue='Campaign name', data=df_filtered_non_zero)
plt.title('Unique Leads (mayores a 0) para las 5 Campa√±as con m√°s Registros a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Unique Leads')
plt.xticks(rotation=45)

# Ajustar la leyenda para que aparezca en la parte inferior
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=1)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Asegurarnos de que 'Reach' y 'Link clicks' sean num√©ricos
df['Reach'] = pd.to_numeric(df['Reach'], errors='coerce')
df['Link clicks'] = pd.to_numeric(df['Link clicks'], errors='coerce')

# Verificar si la columna 'Campaign type' es categ√≥rica, sino convertirla
df['Campaign type'] = df['Campaign type'].astype('category')

# Agrupar los datos por tipo de campa√±a y sumar Reach y Link clicks
df_grouped = df.groupby('Campaign type').agg({'Reach': 'sum', 'Link clicks': 'sum'}).reset_index()

# Definir el ancho de las barras
bar_width = 0.35

# Crear posiciones en el eje X
x = np.arange(len(df_grouped['Campaign type']))

# Crear una figura
plt.figure(figsize=(10, 6))

# Graficar el Reach sumado para cada tipo de campa√±a (barras azules)
plt.bar(x - bar_width/2, df_grouped['Reach'], width=bar_width, color='blue', label='Reach')

# Graficar los Link clicks sumados para cada tipo de campa√±a (barras verdes)
plt.bar(x + bar_width/2, df_grouped['Link clicks'], width=bar_width, color='green', label='Link clicks')

# Configurar etiquetas y t√≠tulo
plt.title('Total de Personas Alcanzadas (Reach) y Clics en Enlace por Tipo de Campa√±a')
plt.xlabel('Tipo de Campa√±a')
plt.ylabel('Personas Alcanzadas (en decenas de millones)')
plt.xticks(ticks=x, labels=df_grouped['Campaign type'], rotation=45)

# Mostrar la leyenda
plt.legend(loc='upper right')

# Mostrar el gr√°fico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar los registros por cada campa√±a
campaign_counts = df['Campaign name'].value_counts()

# Calcular el tercer cuartil (Q3) de los registros por campa√±a
q3_campaign_counts = campaign_counts.quantile(0.75)

# Filtrar las campa√±as que tienen un n√∫mero de registros mayor al tercer cuartil (Q3)
top_campaigns_above_q3 = campaign_counts[campaign_counts > q3_campaign_counts].index

# Filtrar el DataFrame para incluir solo las campa√±as con m√°s registros (mayores a Q3)
df_filtered_q3 = df[df['Campaign name'].isin(top_campaigns_above_q3)]

# Graficar Unique Leads para las campa√±as que tienen m√°s registros (mayores a Q3) a lo largo del tiempo
plt.figure(figsize=(12, 8))
sns.lineplot(x='Date', y='Unique leads', hue='Campaign name', data=df_filtered_q3)
plt.title('Unique Leads para las Campa√±as con m√°s Registros (mayores a Q3) a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('N√∫mero de Leads')
plt.xticks(rotation=45)

# Ajustar la leyenda para que aparezca en la parte inferior
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=1)

plt.tight_layout()
plt.show()
```
